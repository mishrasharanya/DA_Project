```python
# 1) finding outliers using IQR for column total_weekly_hours_worked
q1,q3=df.total_weekly_hours_worked.quantile([0.25,0.75])
IQR=q3-q1
lower,upper=q1-(1.5*IQR),q3+(1.5*IQR)
```

```python
stats_significant_data=df[(df.total_weekly_hours_worked>=lower)&(df.total_weekly_hours_worked<=upper)]
outlying_data=df[(df.total_weekly_hours_worked<lower)|(df.total_weekly_hours_worked>upper)]
outlying_data.index.values
```

```python
print("Lower Limit:",lower)
print("Upper Limit:",upper)
print("Statistacally significant data:")
stats_significant_data
```
