```python
print("Lower Limit:",lower)
print("Upper Limit:",upper)
print("Statistacally significant data:")
stats_significant_data1
```

**What is labour Dependancy ratio?**
The dependency ratio is the total number of people too young or old to work, divided by the number of working-age people (15â€“64 years old). The dependency ratio measures the burden caused by non-working people on a nation's working-age population.



**ANALYSIS:**
Age Dependency Ratios are often used to measure the financial pressure on the actively working population of a community. The higher the ratio, the greater the burden is carried by working-age people. Lower ratios indicate more people are working who can support the dependent population. 
The top 10 countries on the left graph are significantly doing better than the countries on the right graph because the working population has to support less number of people for the right graph. Hence, the per capita income of those countries are higher.
